# 1. Contexto

A indústria de jogos eletrônicos passou de um nicho de entretenimento nas décadas de 1970 e 1980 para se tornar o maior segmento da indústria global de entretenimento, superando em receita tanto o cinema quanto a música combinados. Em 2023, o mercado global de games foi estimado em mais de US$ 180 bilhões, impulsionado pela diversificação de plataformas - de consoles domésticos e PCs a dispositivos móveis - e pela expansão para novos modelos de negócio como jogos free-to-play e microtransações.

Paralelamente ao crescimento comercial, um fenômeno que começou em pequenas competições de LAN nos anos 1990 transformou-se em uma indústria própria: os **e-sports** (esportes eletrônicos). Torneios como *The International* (Dota 2) e o *League of Legends World Championship* passaram a distribuir premiações de dezenas de milhões de dólares, atraindo audiências comparáveis a eventos esportivos tradicionais. A profissionalização dos e-sports trouxe consigo equipes com estrutura corporativa, contratos milionários e ligas organizadas por grandes publishers.

Acompanhando essa transformação, plataformas de streaming ao vivo como a **Twitch** emergiram como o principal canal de consumo de conteúdo de gaming. Com milhões de espectadores simultâneos assistindo a partidas competitivas e conteúdo casual, a Twitch se tornou tanto um termômetro da popularidade de jogos quanto um motor de crescimento para a cena competitiva - criando um ciclo de retroalimentação entre vendas, competição e audiência.

Este trabalho busca explorar as conexões entre essas três dimensões - vendas de jogos, premiações de e-sports e audiência de streaming - através de uma análise de dados que atravessa quatro décadas de evolução da indústria de gaming.

# 2. Descrição do Problema

Nas últimas cinco décadas, a indústria de jogos eletrônicos atravessou transformações profundas - de um mercado nascente dominado por arcades e consoles rudimentares nos anos 1970 para um ecossistema global e multiplataforma que engloba vendas digitais, competições profissionais e streaming ao vivo. Compreender como essas dimensões se conectam e se influenciam mutuamente é fundamental para mapear a trajetória de uma das indústrias de entretenimento mais dinâmicas do mundo.

Este cenário levanta questões que orientam este trabalho:

- **Como as vendas de jogos evoluíram ao longo das décadas, e quais plataformas e gêneros lideraram cada período?**
- **Existe relação entre o crescimento das premiações de e-sports e a popularidade de um jogo na audiência de streaming?**
- **Quais gêneros dominam tanto em vendas comerciais quanto na cena competitiva e no streaming?**

Para investigar essas questões, este trabalho combina três fontes de dados complementares que, juntas, permitem reconstruir a evolução da indústria sob múltiplas perspectivas:

1. **Video Game Sales 2024 (VGChartz):** Dados de vendas globais de 64.016 jogos lançados entre 1971 e 2024, com detalhamento por região (América do Norte, Europa/PAL, Japão, Outros), plataforma, gênero, publisher, developer e critic score - oferecendo uma visão panorâmica do crescimento comercial da indústria ao longo de cinco décadas e 81 plataformas diferentes.

2. **E-Sports Earnings (1998–2024):** Dois datasets complementares que documentam a evolução dos e-sports: um com dados agregados por jogo (669 jogos, incluindo premiação total, número de jogadores e torneios) e outro com dados históricos mensais (10.239 registros de premiações distribuídas em torneios desde 1998) - permitindo analisar o crescimento e a concentração da cena competitiva.

3. **Top Games on Twitch (2016–2024):** Dois datasets que capturam a audiência de gaming no streaming: um com dados mensais dos top jogos na Twitch (21.000 registros, incluindo horas assistidas, pico de viewers, streamers e canais) e outro com métricas globais agregadas da plataforma (105 registros mensais) - documentando a explosão do streaming como canal de consumo de conteúdo de gaming.

A combinação dessas três fontes possibilita uma análise que conecta o sucesso comercial de um jogo ao seu impacto na cena competitiva e na audiência de streaming - revelando padrões que nenhum dataset individual permitiria identificar. Os três datasets compartilham 1.420 jogos em comum entre vendas e Twitch, 394 entre vendas e e-sports, e 181 jogos presentes em todas as três fontes simultaneamente.

## 2.1 Hipóteses do Problema

As hipóteses que orientam esta análise exploratória são:

1. **Migração de plataformas:** As vendas de jogos migraram progressivamente de consoles domésticos para PC e plataformas portáteis/mobile ao longo das décadas, refletindo mudanças no comportamento dos consumidores e na acessibilidade das plataformas.

2. **E-sports como motor de longevidade:** Jogos com cena competitiva de e-sports ativa mantêm audiência elevada na Twitch por períodos mais longos do que jogos sem presença significativa em torneios, sugerindo que a competição profissional funciona como um motor de retenção de público.

3. **Correlação premiação × audiência:** Jogos com maiores premiações acumuladas em e-sports tendem a apresentar maior volume de horas assistidas na Twitch, indicando que investimento em cena competitiva se traduz em engajamento de audiência.

4. **Dominância de gênero:** Gêneros como FPS (First-Person Shooter) e MOBA (Multiplayer Online Battle Arena) dominam simultaneamente as premiações de e-sports e as horas assistidas no streaming, enquanto gêneros tradicionalmente fortes em vendas (como Sports e Platform) têm presença reduzida na cena competitiva.

## 2.2 Tipo de Problema

Este trabalho é primariamente uma **análise exploratória de dados (EDA)** com foco em pré-processamento, combinando análise temporal, correlação entre variáveis e comparações entre categorias.

Os dados permitem enquadrar futuramente diversos tipos de problemas preditivos:

- **Regressão (Video Game Sales):** Prever vendas globais de um jogo com base em plataforma, gênero, publisher e região.
- **Regressão (E-Sports × Twitch):** Prever horas assistidas na Twitch a partir de métricas de e-sports (premiação, número de torneios, jogadores).
- **Classificação:** Prever se um jogo terá presença significativa em e-sports com base em suas características (gênero, plataforma, vendas).
- **Séries temporais:** Modelar a evolução temporal de premiações de e-sports e audiência na Twitch.
- **Clustering:** Agrupar jogos por perfil de sucesso (vendas × e-sports × streaming).

Para fins deste MVP, o foco será na exploração, entendimento e preparação dos dados, deixando a modelagem preditiva como próximo passo natural.

## 5.3 Engenharia de Features

### 5.3.1 Features Derivadas - Video Game Sales

```python
# ============================================================
# Features derivadas - Video Game Sales
# ============================================================

# Década de lançamento
df_vg_sales['decade'] = (df_vg_sales['year'] // 10 * 10).astype('Int64').astype(str) + 's'

# Classificação de plataforma por tipo
console_map = {
    'PS': 'Console', 'PS2': 'Console', 'PS3': 'Console', 'PS4': 'Console', 'PS5': 'Console',
    'XB': 'Console', 'X360': 'Console', 'XOne': 'Console', 'XS': 'Console',
    'Wii': 'Console', 'WiiU': 'Console', 'NS': 'Console',
    'NES': 'Console', 'SNES': 'Console', 'N64': 'Console', 'GC': 'Console',
    'GB': 'Portable', 'GBA': 'Portable', 'DS': 'Portable', '3DS': 'Portable', 'PSP': 'Portable', 'PSV': 'Portable',
    'PC': 'PC',
}
df_vg_sales['platform_type'] = df_vg_sales['console'].map(console_map).fillna('Other')

# Proporção de vendas por região
df_vg_sales['na_share'] = df_vg_sales['na_sales'] / df_vg_sales['total_sales']
df_vg_sales['jp_share'] = df_vg_sales['jp_sales'] / df_vg_sales['total_sales']
df_vg_sales['pal_share'] = df_vg_sales['pal_sales'] / df_vg_sales['total_sales']

print("Features derivadas - Video Game Sales:")
print(f"  decade:        {df_vg_sales['decade'].nunique()} décadas únicas")
print(f"  platform_type: {df_vg_sales['platform_type'].value_counts().to_dict()}")
print(f"  na_share:      média {df_vg_sales['na_share'].mean():.2%}")
```

### 5.3.2 Features Derivadas - E-Sports

```python
# ============================================================
# Features derivadas - E-Sports (usando df_esp_active: jogos > US$ 1M)
# ============================================================

# Premiação média por torneio
df_esp_active['avg_earning_per_tournament'] = (
    df_esp_active['TotalEarnings'] / df_esp_active['TotalTournaments']
)

# Premiação média por jogador
df_esp_active['avg_earning_per_player'] = (
    df_esp_active['TotalEarnings'] / df_esp_active['TotalPlayers']
)

# Idade do jogo no cenário competitivo (anos desde lançamento)
df_esp_active['years_active'] = 2024 - df_esp_active['ReleaseDate']

print("Features derivadas - E-Sports (112 jogos com premiação > US$ 1M):")
print(f"  avg_earning_per_tournament: mediana ${df_esp_active['avg_earning_per_tournament'].median():,.0f}")
print(f"  avg_earning_per_player:     mediana ${df_esp_active['avg_earning_per_player'].median():,.0f}")
print(f"  years_active:               mediana {df_esp_active['years_active'].median():.0f} anos")
```

```python
# ============================================================
# Features derivadas - E-Sports Historical
# ============================================================

# Filtrar historical apenas para jogos presentes no df_esp_active
active_games = df_esp_active['Game'].unique()
df_esp_history_active = df_esp_history[df_esp_history['Game'].isin(active_games)].copy()

# Premiação acumulada (running total por jogo)
df_esp_history_active = df_esp_history_active.sort_values(['Game', 'Date'])
df_esp_history_active['cumulative_earnings'] = df_esp_history_active.groupby('Game')['Earnings'].cumsum()

print(f"E-Sports Historical filtrado: {len(df_esp_history_active):,} registros ({len(active_games)} jogos)")
print(f"  cumulative_earnings adicionado (running total por jogo)")
```

### 5.3.3 Features Derivadas - Twitch

```python
# ============================================================
# Features derivadas - Twitch
# ============================================================

# Eficiência de audiência: horas assistidas por streamer
df_tw_games['hours_per_streamer'] = df_tw_games['Hours_watched'] / df_tw_games['Streamers']

# Proporção de pico: Peak_viewers / Avg_viewers (mede "spikiness" da audiência)
df_tw_games['peak_ratio'] = df_tw_games['Peak_viewers'] / df_tw_games['Avg_viewers']

# Agregar totais por jogo (all-time) para cruzamentos
df_twitch_totals = (df_tw_games
    .groupby('Game')
    .agg(
        total_hours_watched=('Hours_watched', 'sum'),
        total_hours_streamed=('Hours_streamed', 'sum'),
        avg_viewer_ratio_mean=('Avg_viewer_ratio', 'mean'),
        total_streamers=('Streamers', 'sum'),
        months_active=('Date', 'nunique'),
    )
    .reset_index()
)

print("Features derivadas - Twitch Game Data:")
print(f"  hours_per_streamer: mediana {df_tw_games['hours_per_streamer'].median():,.0f}")
print(f"  peak_ratio:         mediana {df_tw_games['peak_ratio'].median():,.1f}")
print(f"\nTabela agregada por jogo (df_twitch_totals): {len(df_twitch_totals):,} jogos")
```

## 6.1 Hipótese 1 - Migração de Plataformas

**Hipótese:** As vendas de jogos migraram progressivamente de consoles domésticos para PC e plataformas portáteis/mobile ao longo das décadas.

```python
# ============================================================
# Hipótese 1: Evolução da participação por tipo de plataforma
# ============================================================

# Filtrar registros sem década válida e remover 2020s (dados incompletos)
df_vg_h1 = df_vg_sales[
    df_vg_sales['decade'].notna() &
    (df_vg_sales['decade'] != '<NA>s') &
    (df_vg_sales['decade'] != '2020s')
].copy()

sales_by_type = df_vg_h1.groupby(['decade', 'platform_type'])['total_sales'].sum().unstack(fill_value=0)
sales_pct = sales_by_type.div(sales_by_type.sum(axis=1), axis=0) * 100

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sales_by_type.plot(kind='bar', stacked=True, ax=axes[0],
                   color=PALETTE_CATEGORICAL[:len(sales_by_type.columns)],
                   edgecolor='white', alpha=0.85)
axes[0].set_xlabel('Decada')
axes[0].set_ylabel('Vendas Totais (milhoes de unidades)')
axes[0].set_title('Vendas por Tipo de Plataforma (absoluto)')
axes[0].tick_params(axis='x', rotation=0)
axes[0].legend_.remove()

sales_pct.plot(kind='bar', stacked=True, ax=axes[1],
               color=PALETTE_CATEGORICAL[:len(sales_pct.columns)],
               edgecolor='white', alpha=0.85)
axes[1].set_xlabel('Decada')
axes[1].set_ylabel('Participacao (%)')
axes[1].set_title('Participacao por Tipo de Plataforma (%)')
axes[1].tick_params(axis='x', rotation=0)
axes[1].legend(title='Tipo', bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=9)

add_note(axes[0], 'Nota: decada de 2020 excluida por falta\nde cobertura de vendas no VGChartz')

plt.suptitle('Hipotese 1: Migracao de Plataformas ao Longo das Decadas',
             fontsize=14, fontweight='bold', color=COLORS['neutral'])
plt.tight_layout()
plt.show()

print("\nParticipacao percentual por decada:\n")
print(sales_pct.round(1).to_string())
```

**Conclusao - Hipotese 1: Parcialmente confirmada (com limitacoes de dados).**

A analise revela que **consoles domesticos dominaram consistentemente as vendas** entre as decadas de 1990 e 2010, mantendo participacao entre 74% e 85%. A hipotese de migracao para PC nao se confirma nos dados de vendas - PC nunca ultrapassou 4% do mercado por volume de unidades vendidas.

A hipotese se confirma parcialmente para **plataformas portateis**, que cresceram de 1,7% nos anos 1990 para 23,4% nos anos 2000 (era DS e PSP), recuando para 12,6% nos anos 2010.

A decada de 2020 foi excluida desta analise por falta de cobertura: o VGChartz nao possui dados de vendas para PS5 e Xbox Series, e os poucos registros de 2020 sao relancamentos de titulos antigos. Alem disso, o dataset foca em vendas de midia fisica e consoles tradicionais, **subrepresentando significativamente o mercado de PC digital** (Steam, Epic Games) e **mobile** (App Store, Google Play), que juntos representam a maioria da receita global de gaming atual. Portanto, a dominancia de consoles observada reflete mais a cobertura da fonte do que necessariamente a realidade completa do mercado.

## 6.2 Hipotese 2 - E-Sports como Motor de Longevidade

**Hipotese:** Jogos com cena competitiva de e-sports ativa mantêm audiencia elevada na Twitch por periodos mais longos do que jogos sem presenca significativa em torneios.

```python
# ============================================================
# Hipotese 2: Longevidade na Twitch - e-sports vs. nao e-sports
# ============================================================

# Meses ativos e horas totais por jogo na Twitch
twitch_longevity = (df_tw_games
    .groupby('Game')
    .agg(
        months_active=('Date', 'nunique'),
        total_hours=('Hours_watched', 'sum'),
        avg_monthly_hours=('Hours_watched', 'mean'),
    )
    .reset_index()
)

esport_games = set(df_esp_active['Game'].str.lower().str.strip())
twitch_longevity['is_esport'] = twitch_longevity['Game'].str.lower().str.strip().isin(esport_games)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

data_months = [
    twitch_longevity[twitch_longevity['is_esport']]['months_active'],
    twitch_longevity[~twitch_longevity['is_esport']]['months_active']
]
bp1 = axes[0].boxplot(data_months, labels=['E-Sports\n(premiacao > US$ 1M)', 'Sem E-Sports'],
                       patch_artist=True, medianprops=dict(color=COLORS['accent'], linewidth=2))
bp1['boxes'][0].set_facecolor(COLORS['primary'])
bp1['boxes'][0].set_alpha(0.7)
bp1['boxes'][1].set_facecolor(COLORS['primary_pale'])
axes[0].set_ylabel('Meses Ativos na Twitch')
axes[0].set_title('Longevidade na Twitch')

data_hours = [
    twitch_longevity[twitch_longevity['is_esport']]['avg_monthly_hours'],
    twitch_longevity[~twitch_longevity['is_esport']]['avg_monthly_hours']
]
bp2 = axes[1].boxplot(data_hours, labels=['E-Sports\n(premiacao > US$ 1M)', 'Sem E-Sports'],
                       patch_artist=True, medianprops=dict(color=COLORS['accent'], linewidth=2))
bp2['boxes'][0].set_facecolor(COLORS['primary'])
bp2['boxes'][0].set_alpha(0.7)
bp2['boxes'][1].set_facecolor(COLORS['primary_pale'])
axes[1].set_ylabel('Horas Assistidas Mensais (media)')
axes[1].set_title('Audiencia Media Mensal na Twitch')
axes[1].set_yscale('log')

plt.suptitle('Hipotese 2: E-Sports como Motor de Longevidade',
             fontsize=14, fontweight='bold', color=COLORS['neutral'])
plt.tight_layout()
plt.show()

esport_stats = twitch_longevity[twitch_longevity['is_esport']]
non_esport_stats = twitch_longevity[~twitch_longevity['is_esport']]

print(f"{'Metrica':<30} {'E-Sports (mediana)':>20} {'Sem E-Sports (mediana)':>25}")
print("-" * 77)
print(f"{'Meses ativos na Twitch':<30} {esport_stats['months_active'].median():>20.0f} {non_esport_stats['months_active'].median():>25.0f}")
print(f"{'Horas totais assistidas':<30} {esport_stats['total_hours'].median():>20,.0f} {non_esport_stats['total_hours'].median():>25,.0f}")
print(f"{'Media mensal de horas':<30} {esport_stats['avg_monthly_hours'].median():>20,.0f} {non_esport_stats['avg_monthly_hours'].median():>25,.0f}")
```

**Conclusao - Hipotese 2: Fortemente confirmada.**

A diferenca de longevidade entre jogos de e-sports e jogos sem cena competitiva é dramatica:

- Jogos de e-sports permanecem ativos na Twitch por uma **mediana de 26 meses**, contra apenas **2 meses** para jogos sem e-sports - uma diferenca de 13x.
- Em volume de audiencia, jogos de e-sports acumulam uma **mediana de aproximadamente 84 milhoes de horas** assistidas, contra aproximadamente **1,5 milhao** para os demais - uma diferenca de aproximadamente 57x.

Esses dados sustentam fortemente a hipotese de que a cena competitiva funciona como um **motor de retencao de audiencia**: torneios regulares, ligas organizadas e a narrativa de competicao profissional mantêm o interesse do publico muito alem do ciclo de vida natural de um jogo. Titulos como League of Legends, CS:GO e Dota 2 sao exemplos claros desse fenomeno, mantendo audiencias massivas por anos apos seus lancamentos.

## 6.3 Hipotese 3 - Correlacao Premiacao x Audiencia

**Hipotese:** Jogos com maiores premiacoes acumuladas em e-sports tendem a apresentar maior volume de horas assistidas na Twitch.

```python
# ============================================================
# Hipotese 3: Correlacao entre premiacao e audiencia
# ============================================================

# Merge: e-sports (>1M) x Twitch totals
merged_h3 = df_esp_active.merge(df_twitch_totals, on='Game', how='inner')
print(f"Jogos presentes em ambos os datasets: {len(merged_h3)}")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

axes[0].scatter(merged_h3['TotalEarnings'] / 1e6,
               merged_h3['total_hours_watched'] / 1e9,
               s=merged_h3['TotalTournaments'] * 0.3,
               alpha=0.5, c=COLORS['primary'], edgecolors='white', linewidth=0.5)

for _, row in merged_h3.nlargest(6, 'TotalEarnings').iterrows():
    axes[0].annotate(row['Game'],
                     (row['TotalEarnings'] / 1e6, row['total_hours_watched'] / 1e9),
                     fontsize=9, alpha=0.8, color=COLORS['neutral'],
                     xytext=(5, 5), textcoords='offset points')

axes[0].set_xlabel('Premiacao Total E-Sports (milhoes USD)')
axes[0].set_ylabel('Horas Assistidas Twitch (bilhoes)')
axes[0].set_title('Escala Linear')

axes[1].scatter(merged_h3['TotalEarnings'],
               merged_h3['total_hours_watched'],
               s=merged_h3['TotalTournaments'] * 0.3,
               alpha=0.5, c=COLORS['primary_light'], edgecolors='white', linewidth=0.5)

for _, row in merged_h3.nlargest(6, 'TotalEarnings').iterrows():
    axes[1].annotate(row['Game'],
                     (row['TotalEarnings'], row['total_hours_watched']),
                     fontsize=9, alpha=0.8, color=COLORS['neutral'],
                     xytext=(5, 5), textcoords='offset points')

axes[1].set_xscale('log')
axes[1].set_yscale('log')
axes[1].set_xlabel('Premiacao Total E-Sports (USD, log)')
axes[1].set_ylabel('Horas Assistidas Twitch (log)')
axes[1].set_title('Escala Log-Log')

plt.suptitle('Hipotese 3: Premiacao de E-Sports vs. Audiencia na Twitch\n(tamanho = numero de torneios)',
             fontsize=14, fontweight='bold', color=COLORS['neutral'])
plt.tight_layout()
plt.show()

corr_linear = merged_h3[['TotalEarnings', 'total_hours_watched']].corr().iloc[0, 1]
corr_log = np.log10(merged_h3['TotalEarnings']).corr(np.log10(merged_h3['total_hours_watched']))
print(f"\nCorrelacao de Pearson (linear):  {corr_linear:.3f}")
print(f"Correlacao de Pearson (log-log): {corr_log:.3f}")
```

**Conclusao - Hipotese 3: Confirmada com ressalvas.**

A analise revela correlacao positiva moderada entre premiacao de e-sports e audiencia na Twitch:

- **Correlacao linear (Pearson): 0.638** - indica que jogos com maiores premiacoes tendem a ter mais horas assistidas.
- **Correlacao log-log: 0.585** - confirma que a relacao se mantem mesmo quando controlamos para a enorme variacao de escala entre os jogos.

No entanto, a correlacao nao é perfeita, e o scatter plot revela casos interessantes:

- **Dota 2** tem a maior premiacao da historia (aproximadamente US\$ 360M) mas nao é o jogo mais assistido na Twitch - League of Legends lidera em audiencia com uma premiacao significativamente menor (aproximadamente US\$ 114M). Isso sugere que **modelo de distribuicao de premiacoes** (The International concentra tudo em um unico torneio) nao é o unico fator de audiencia.
- Jogos como **GTA V** e **Minecraft** (quando presentes no merge) possuem alta audiencia na Twitch sem premiacoes expressivas de e-sports, demonstrando que o streaming vai alem da cena competitiva.

A relacao entre premiacao e audiencia é real, mas outros fatores - como acessibilidade do jogo, qualidade do conteúdo de streaming e personalidade dos streamers - tambem influenciam significativamente a audiencia.

## 6.4 Hipotese 4 - Dominancia de Genero

**Hipotese:** Generos como FPS e MOBA dominam simultaneamente as premiacoes de e-sports e as horas assistidas no streaming, enquanto generos fortes em vendas (como Sports e Platform) têm presenca reduzida na cena competitiva.

```python
# ============================================================
# Hipotese 4: Comparacao de generos entre as 3 dimensoes
# ============================================================

# E-Sports: premiação por gênero
esports_genre = df_esp_active.groupby('Genre')['TotalEarnings'].sum().nlargest(8)

# Twitch: horas por gênero (via merge com e-sports para ter o gênero)
merged_genre = df_esp_active[['Game', 'Genre']].merge(df_twitch_totals, on='Game', how='inner')
twitch_genre = merged_genre.groupby('Genre')['total_hours_watched'].sum().nlargest(8)

# VG Sales: vendas por gênero
vg_genre = df_vg_sales.groupby('genre')['total_sales'].sum().nlargest(8)

fig, axes = plt.subplots(1, 3, figsize=(20, 7))

# E-Sports
esports_genre.sort_values().plot(kind='barh', ax=axes[0],
    color=COLORS['primary'], edgecolor='white', alpha=0.85)
axes[0].set_xlabel('Premiacao Total (USD)')
axes[0].set_title('E-Sports\n(por premiacao)')
axes[0].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'${x/1e6:.0f}M'))

# Twitch
twitch_genre.sort_values().plot(kind='barh', ax=axes[1],
    color=COLORS['primary_light'], edgecolor='white', alpha=0.85)
axes[1].set_xlabel('Horas Assistidas')
axes[1].set_title('Twitch\n(por horas assistidas)')
axes[1].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x/1e9:.1f}B'))

# VG Sales
vg_genre.sort_values().plot(kind='barh', ax=axes[2],
    color=COLORS['primary_pale'], edgecolor='white', alpha=0.85)
axes[2].set_xlabel('Vendas Globais (milhoes unidades)')
axes[2].set_title('Video Game Sales\n(por vendas)')

plt.suptitle('Hipotese 4: Dominancia de Genero - E-Sports vs. Twitch vs. Vendas',
             fontsize=14, fontweight='bold', color=COLORS['neutral'])
plt.tight_layout()
plt.show()
```

**Conclusao - Hipotese 4: Confirmada.**

A comparacao entre as tres dimensoes revela perfis de genero marcadamente diferentes:

**E-Sports e Twitch sao dominados pelos mesmos generos:**
- **MOBA** (Multiplayer Online Battle Arena) lidera tanto em premiacao (aproximadamente US\$ 641M) quanto em horas assistidas (aproximadamente 19,7B horas), impulsionado por League of Legends e Dota 2.
- **FPS** (First-Person Shooter) ocupa a segunda posicao em ambas as dimensoes (aproximadamente US\$ 471M e 16B horas), com CS:GO/CS2, VALORANT e Rainbow Six Siege como pilares.
- **Battle Royale** completa o top 3 (aproximadamente US\$ 389M e 8,9B horas), liderado por Fortnite e PUBG.

**Vendas comerciais têm perfil completamente diferente:**
- **Sports** e **Action** lideram em vendas globais, generos que têm presenca modesta em e-sports.
- **Shooter** aparece no top 3 de vendas, servindo como ponte entre o mercado comercial e competitivo.
- Generos como **Platform** e **Racing**, fortes em vendas, sao praticamente irrelevantes na cena de e-sports.

Essa divergencia confirma que **sucesso comercial e sucesso competitivo sao dimensoes distintas** da industria de gaming. Generos competitivos (MOBA, FPS, Battle Royale) sao projetados para engajamento de longo prazo e espectatoriedade, enquanto generos comerciais (Sports, Action, Platform) priorizam experiencias single-player ou casual que nao se traduzem naturalmente em audiencia competitiva.

# 7. Conclusão

Este trabalho realizou uma análise exploratória e pré-processamento de cinco datasets complementares que, juntos, cobrem três dimensões da indústria de gaming: vendas comerciais (1971–2024), premiações de e-sports (1998–2024) e audiência de streaming na Twitch (2016–2024).

**Principais achados:**

A indústria de games apresentou crescimento consistente ao longo de quatro décadas, com consoles domésticos dominando as vendas entre as décadas de 1990 e 2010 - representando entre 74% e 85% do mercado por volume de unidades vendidas. Plataformas portáteis tiveram seu auge nos anos 2000 com o Nintendo DS e PSP, enquanto o mercado de PC permaneceu subrepresentado nos dados de vendas físicas, embora seja líder em volume de lançamentos.

Os gêneros MOBA, FPS e Battle Royale dominam amplamente a cena competitiva, com perfil marcadamente diferente dos gêneros que lideram em vendas comerciais (Sports, Action).

Em volume de audiência, essa diferença chega a 57 vezes. **Limitações identificadas:**

- O dataset VGChartz foca em vendas de mídia física, subrepresentando o mercado digital (Steam, Epic Games) e mobile, que juntos constituem a maioria da receita global atual. A geração atual de consoles (PS5, Xbox Series) não possui dados de vendas.

- Os dados da Twitch começam em 2016, limitando comparações temporais com as outras fontes.
- Os dados de e-sports de 2024 cobrem apenas janeiro a outubro, e mesmo anualizados indicam uma retração de aproximadamente 24% em relação a 2023, sugerindo um período de ajuste na indústria.

Apesar dessas limitações, a combinação dos três datasets permitiu uma análise multidimensional que revelou padrões que nenhuma fonte individual seria capaz de mostrar - conectando o sucesso comercial de um jogo ao seu impacto competitivo e à sua audiência de streaming.
